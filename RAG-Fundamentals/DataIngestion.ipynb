### Data Ingestion using Langchain Docmument Loaders

```python
import os
```

```python
from langchain_community.document_loaders import TextLoader
from langchain_community.document_loaders import PyPDFLoader
from langchain_community.document_loaders import WebBaseLoader
```

### Loading Text Documents

```python
FILE_PATH = os.path.join(os.getcwd(), 'Files', 'speech.txt')

loader = TextLoader(FILE_PATH)
documents = loader.load()
```

```python
documents
```

### Loading PDF Documents

```python
FILE_PATH = os.path.join(os.getcwd(), 'Files', 'attention.pdf')

loader = PyPDFLoader(FILE_PATH)
documents = loader.load()
```

```python
documents
```

### Loading Documents From Web

```python
web_paths = [
    'https://python.langchain.com/docs/introduction/',
    'https://docs.smith.langchain.com/',
    'https://www.ibm.com/think/topics/ai-agents'
]

loader = WebBaseLoader(web_paths=web_paths)
documents = loader.load()
```

```python
documents
```

```python
documents[0].metadata['source']
```

### Loading Documents with WikipediaLoader

```python
from langchain_community.document_loaders import WikipediaLoader
```

```python
loader = WikipediaLoader(query='Generative AI', load_max_docs=2)
documents = loader.load()
```
